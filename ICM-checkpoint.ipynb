```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

Now we have to load the data into numpy arrays....

```python
data_train_path = 'Fruits_Vegetables/train'
data_test_path = 'Fruits_Vegetables/test'
data_val_path = 'Fruits_Vegetables/validation'
```

Now we have to set the size of the imagesss......

```python
img_width = 180
img_height =180
```

Now we have to bring the data in the form of arrays....

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle=True,
    image_size=(img_width, img_height),
    batch_size=32,
    validation_split=False)
```

Now we have to check different types of fruit and vegetables we have

```python
data_cat = data_train.class_names
```

```python
data_val = tf.keras.utils.image_dataset_from_directory(data_val_path,
                                                       image_size=(img_height,img_width),
                                                       batch_size=32,
                                                        shuffle=False,
                                                       validation_split=False)
```

```python
data_cat = data_val.class_names
```

```python
from tensorflow.keras.models import Sequential
```

```python
data_train
```

```python
model = Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32,3, padding='same',activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dropout(0.2),
    layers.Dense(128),
    layers.Dense(len(data_cat))                
])
```

This code defines a convolutional neural network (CNN) model using Keras' Sequential API.

1. `layers.Rescaling(1./255)`: This layer rescales the input pixel values to be in the range [0, 1], which is a common preprocessing step for image data.

2. `layers.Conv2D(16, 3, padding='same', activation='relu')`: This is a convolutional layer with 16 filters (output channels), each of size 3x3 pixels. The 'same' padding ensures that the output has the same spatial dimensions as the input, and 'relu' is the activation function used to introduce non-linearity.

3. `layers.MaxPooling2D()`: This layer performs max pooling, which reduces the spatial dimensions of the feature maps by taking the maximum value in each pooling window. No arguments are provided, so it defaults to a 2x2 pooling window with a stride of 2.

4. This pattern of convolutional and max pooling layers repeats twice more, gradually increasing the number of filters in each convolutional layer (32 and then 64) while downsampling the spatial dimensions through max pooling.

5. `layers.Flatten()`: This layer flattens the 2D feature maps into a 1D vector, preparing them for input into a fully connected (dense) layer.

6. `layers.Dropout(0.2)`: This layer applies dropout regularization, randomly setting a fraction of input units to 0 during training to prevent overfitting. Here, 20% of the input units are randomly dropped out.

7. `layers.Dense(128)`: This is a fully connected (dense) layer with 128 neurons and no activation function specified, which means it defaults to a linear activation function.

8. `layers.Dense(len(data_cat))`: This is the output layer of the network, also a fully connected layer. The number of neurons matches the number of categories/classes in the dataset (represented by `len(data_cat)`), and it typically uses a softmax activation function for multi-class classification tasks.

This model architecture is commonly used for image classification tasks, with convolutional layers extracting hierarchical features from the input images and dense layers performing classification based on these features.

```python
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```



1. `optimizer='adam'`: This specifies the optimizer to be used during training. In this case, 'adam' refers to the Adam optimizer, which is an adaptive learning rate optimization algorithm known for its efficiency and effectiveness.

2. `loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)`: This specifies the loss function to be used for training. Here, `tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)` indicates that we're using the sparse categorical crossentropy loss, which is commonly used for multi-class classification tasks when the labels are provided as integers. The `from_logits=True` argument indicates that the model outputs logits (raw scores) instead of probabilities, and the loss function will apply the softmax function internally.

3. `metrics=['accuracy']`: This specifies the evaluation metric(s) to be monitored during training and testing. Here, we're using the 'accuracy' metric to track the percentage of correctly classified samples during training and evaluation.

Overall, this line configures the model for training by specifying the optimizer, loss function, and evaluation metric(s) to be used.

```python
epochs_size = 25
history = model.fit(data_train, validation_data=data_val, epochs=epochs_size)
```

```python
epochs_range = range(epochs_size)
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.plot(epochs_range,history.history['accuracy'],label = 'Training Accuracy')
plt.plot(epochs_range, history.history['val_accuracy'],label = 'Validation Accuracy')
plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(epochs_range,history.history['loss'],label = 'Training Loss')
plt.plot(epochs_range, history.history['val_loss'],label = 'Validation Loss')
plt.title('Loss')
```

```python
image = 'corn.jpg'
image = tf.keras.utils.load_img(image, target_size=(img_height,img_width))
img_arr = tf.keras.utils.array_to_img(image)
img_bat=tf.expand_dims(img_arr,0)
```

```python
predict = model.predict(img_bat)
```

```python
score = tf.nn.softmax(predict)
```

1. `print()`: This function is used to display output to the console.

2. `'Veg/Fruit in image is {} with accuracy of {:0.2f}'`: This is a formatted string that serves as the template for the output. It contains two placeholders (`{}` and `{}`) that will be replaced with actual values.

3. `.format()`: This method is used to insert values into the formatted string. It takes the values provided as arguments and substitutes them into the placeholders in the string.

4. `data_cat[np.argmax(score)]`: This expression retrieves the category (vegetable or fruit) corresponding to the highest score (probability) in the `score` array. `np.argmax(score)` returns the index of the maximum value in the `score` array, which represents the predicted category. `data_cat[]` then retrieves the corresponding category name from the `data_cat` array.

5. `np.max(score) * 100`: This expression calculates the maximum score (probability) in the `score` array and multiplies it by 100 to convert it to a percentage. This represents the accuracy of the prediction.

So, when this line of code is executed, it prints a message indicating the predicted category of the image along with the corresponding accuracy score. ♢

```python
print('Veg/Fruit in image is {} with accuracy of {:0.2f}'.format(data_cat[np.argmax(score)],np.max(score)*100))
```
